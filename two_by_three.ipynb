```python
import numpy as np
import matplotlib.pyplot as plt
from DifferentialSystem import DifferentialSystem
from PayoffMatrices import PayoffMatrices
```

```python
player_one_payoffs = [[20, 43, -260], [-2, 21, 19]]
player_two_payoffs = [[-7.5, -23, -13.8], [-5, 0, -3]]
```

```python
def checker(player_one_payoffs, player_two_payoffs):
    A = player_two_payoffs[0][0] - player_two_payoffs[0][2]
    B = player_two_payoffs[0][2] - player_two_payoffs[0][1]
    C = player_two_payoffs[1][1] - player_two_payoffs[1][2]
    D = player_two_payoffs[1][2] - player_two_payoffs[1][0]
    unique = (player_one_payoffs[0][0] > player_one_payoffs[1][0]) and (player_one_payoffs[0][1] > player_one_payoffs[1][1]) and (player_one_payoffs[0][2] < player_one_payoffs[1][2])
    unique = unique and (A > 0) and (B > 0) and (C > 0) and (D > 0) and (A * C > B * D)
    print(B > A)
    print(C > D)
    print(A * C)
    print(B * D)
    return unique
```

```python
checker(player_one_payoffs, player_two_payoffs)
```

```python
matrices = PayoffMatrices(player_one_payoffs, player_two_payoffs)
print(matrices.player_one_payoffs)
print(matrices.player_two_payoffs)
transition_matrix = matrices.transition_matrix()
system = DifferentialSystem(len(player_one_payoffs), len(player_one_payoffs[0]), transition_matrix)
```

```python
player_one_utilities, player_two_utilities, player_one_strategies, player_two_strategies = system.run([0.495, 0.505, 0.02, 0.98, 0], 500_000, 0.00001)
```

```python
plt.plot(np.transpose(player_one_strategies)[0])
plt.plot(np.transpose(player_one_strategies)[1])
plt.legend(['1', '2'])
```
